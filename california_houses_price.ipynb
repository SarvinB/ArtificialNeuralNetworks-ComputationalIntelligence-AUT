**Read dataset as train dataset and test data set**

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from layers.convolution2d import Conv2D
from layers.maxpooling2d import MaxPool2D
from layers.fullyconnected import FC
import matplotlib.pyplot as plt
from activations import Activation, get_activation
from optimizers.gradientdescent import GD
from losses.meansquarederror import MeanSquaredError
from model import Model
from sklearn.preprocessing import StandardScaler
```

```python
df_train = pd.read_csv('/Users/sarvinbaghi/Desktop/6/Neural Network/project1/ArtificialNeuralNetworks-ComputationalIntelligence-AUT/datasets/california_houses_price/california_housing_train.csv')
df_test = pd.read_csv('/Users/sarvinbaghi/Desktop/6/Neural Network/project1/ArtificialNeuralNetworks-ComputationalIntelligence-AUT/datasets/california_houses_price/california_housing_test.csv')
```

```python
dataset_train = df_train.values
dataset_test = df_test.values
```

**scale dataset**

```python
scaler = StandardScaler()
train_scaled = scaler.fit_transform(dataset_train)
test_scaled = scaler.fit_transform(dataset_test)
```

```python
X_train = train_scaled[:,0:8]
Y_train = train_scaled[:,8:9]+2
X_test = test_scaled[:,0:8]
Y_test = test_scaled[:,8:9]+2
```

**Distribution of features**

```python
stop = 500
fig, axs = plt.subplots(2, 4, figsize=(12, 5))
axs[0][0].plot(X_train.T[0][1:stop], Y_train[1:stop], 'o', color="orange")
axs[0][1].plot(X_train.T[1][1:stop], Y_train[1:stop], 'o', color="red")
axs[0][2].plot(X_train.T[2][1:stop], Y_train[1:stop], 'o', color="blue")
axs[0][3].plot(X_train.T[3][1:stop], Y_train[1:stop], 'o', color="black")
axs[1][0].plot(X_train.T[4][1:stop], Y_train[1:stop], 'o', color="pink")
axs[1][1].plot(X_train.T[5][1:stop], Y_train[1:stop], 'o', color="cyan")
axs[1][2].plot(X_train.T[6][1:stop], Y_train[1:stop], 'o', color="purple")
axs[1][3].plot(X_train.T[7][1:stop], Y_train[1:stop], 'o', color="green")
```

**model has three fully connected layer with relu activation function, GD as optimizer and MSE as loos_function**

```python
layer1 = FC(8, 10, "l1")
activation1 = get_activation("relu")
layer2 = FC(10, 10, "l2")
activation2 = get_activation("relu")
layer3 = FC(10, 1, "l3")
activation3 = get_activation("relu")
learning_rate = 0.01
epochs = 100
arch = {"l1" : layer1, "a1" : activation1, "l2" : layer2, "a2" : activation2, "l3" : layer3, "a3" : activation3}
loos_function = MeanSquaredError()
optimizer = GD(arch, learning_rate)
model = Model(arch, loos_function, optimizer)
train_cost, val_cost = model.train(X_train.T, Y_train.T, epochs, shuffling=True)
```

**Result of training in plot form (Train data as input)**

```python
stop = 50
fig, axs = plt.subplots(1, 1, figsize=(12, 5))
xpoints = np.asanyarray(range(stop))
axs.plot(xpoints, np.asarray(Y_train).reshape((X_train.shape[0], 1))[0:stop], 'o', color="orange")
axs.plot(xpoints, np.asarray(model.forward(X_train.T)[5][0]).reshape((X_train.shape[0], 1))[0:stop], 'o', color="blue")
```

**Result of training in plot form (Test data as input)**

```python
Y_predict = model.forward(X_test.T)
Y_predict_train = model.forward(X_train.T)
stop = 50
fig, axs = plt.subplots(1, 1, figsize=(12, 5))
xpoints = np.asanyarray(range(stop))
axs.plot(xpoints, np.asarray(Y_test).reshape((X_test.shape[0], 1))[0:stop], 'o', color="orange")
axs.plot(xpoints, np.asarray(Y_predict[5][0]).reshape((X_test.shape[0], 1))[0:stop], 'o', color="blue")
print(loos_function.compute(Y_predict[5], Y_test))
print(loos_function.compute(Y_predict_train[5], Y_train))
# print(Y_predict[3][0][0:10])
# print(Y_test[0:10].T)

```
